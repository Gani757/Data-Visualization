stcked bar plot
- x axis for 2 or more plots should be same
- all x-axis in  same sequence

```python
company=["google","amzn","fb","ms"]
turnover= [90,140,88,22]
profit= [40,20,34,12]

plt.bar(company,turnover,label="revenue")
plt.bar(company,profit,label="profit")
plt.legend()
plt.show()
```

🧠 Think of it this way:
label= → Defines what should be in the legend

plt.legend() → Displays the legend on the chart

No legend = Labels are never shown, even though they exist in code
