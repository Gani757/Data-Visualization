```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

****Plots for Continuous variables********
-

univariate
-
- histogram
- box plot

```python
df.columns
```

```python
sns.histplot(data=df,x='total_bill',bins=10,kde=True)
plt.show()
```

box plot

```python
sns.boxplot(df['total_bill'])
plt.show()
```

bivariate:
-
- scatter plot  (relation b/w 2 variables & hue(optinal))
- line plot      (time series)
- joint plot  (scatter plot+ histogram)

```python
sns.scatterplot(data=df,x='total_bill',y="tip",hue="sex")
plt.show()
```

```python
sns.relplot(data=df,x='total_bill',y="tip",size="sex")
plt.show()
```

```python
sns.relplot(data=df,x='total_bill',y="tip",hue="sex",style="smoker")
plt.show()
```

```python
sns.relplot(data=df,x='total_bill',y="tip",hue="sex",col="size",col_wrap=2)
plt.show()
```

```python
# line plot

sns.relplot(data=df,x='total_bill',y="tip",kind="line")
plt.show()
```

```python
sns.jointplot(data=df,x='total_bill',y="tip",hue="sex",kind="scatter")
plt.show()
```

violin plot--(univariate continuous)

```python
sns.violinplot(data=df,x="total_bill",y="day",hue="day")
plt.show()
```

multiple continuous variable
-
- pair plot
- heatmap

```python
sns.pairplot(df,x_vars='total_bill',hue="sex",kind='scatter')
plt.show()
```

```python
c_m=df[["total_bill","tip"]].corr()
c_m
```

```python
sns.heatmap(c_m,annot=True)
```

```python
df["smoker"].value_counts()
```

```python
sns.countplot(data=df,x="smoker",hue="smoker")
plt.show()
```

```python
sns.countplot(data=df,x="smoker",hue="sex")
plt.show()
```

```python
sns.catplot(data=df,x="sex",y="total_bill",hue="sex")
plt.show()
```

```python
sns.catplot(data=df,x="sex",y="total_bill",hue="sex",kind="bar")
plt.show()
```

```python
sns.catplot(data=df,x="day",y="total_bill",hue="size",kind="bar")
plt.show()
```

```python
sns.catplot(data=df,x="smoker",y="total_bill",hue="size",kind="bar")
plt.show()
```
